# Loading Data

```python
import json
import ast 
import pandas as pd
import numpy as np
from IPython import display
data = []
fracture = pd.read_csv("groupedProportionFracture.csv") # Fracture Data
pd.set_option('display.max_rows', 1000)
```

```python
# Loading Chat Data
for _ in set(fracture["batch"]):
    with open('../.data/' + str(_) + "/chats.json", "r") as f:
        chat = json.load(f)
        data.append(chat)
                
```

```python
import numpy as np
def getfracture(row, fracture, batchID):
    f = fracture[(fracture["batch"] == batchID)  & (fracture["room"] == row["team"]) & (fracture["round"] == row["round"])]["fracture"]
    if (list(f)==[0]):
        return 0
    elif (list(f)==[]):
        return np.nan
    else:
        return 1

```

# Splitting Data into Interactions

```python
frame = {}
for _ in data:
    print ("Batch", _[0]["batch"])
    batch = [(chat["round"] + 1, chat["room"], chat["userID"], chat["message"], chat["time"], 0) 
                      for chat in _]
    batch = pd.DataFrame(batch)
    batch.columns = ["round", "team", "userID", "message", "time", "fracture"]
    batch = batch.sort_values(["round", "team", "time"])
    batch = batch[batch["team"] != ""]
    batch["fracture"] = batch.apply(getfracture, args=(fracture, _[0]["batch"]), axis = 1)
    batch.index = np.arange(1, batch.shape[0] + 1)
    for index, row in batch.iterrows():
        identifier = tuple([_[0]["batch"], row["round"], row["team"]])
        if identifier not in frame.keys():
            frame[identifier] = {"message": [row["message"]], "fracture": row["fracture"]}
        else:
            frame[identifier]["message"].append(row["message"])

        
    
```

```python
frame
```

```python
chats = [(val["message"]) for key, val in frame.items() if not np.isnan(val["fracture"])]
fractures = [val["fracture"] for key, val in frame.items() if not np.isnan(val["fracture"])]
print(chats[30])
print(fractures[30])
```

Define a class that will be our bag of words mapper.

```python
class Lang:
    def __init__(self):
        self.word2index = {}
        self.index2word = {}
        self.n_words = 0  # Count SOS and EOS
        self.word2count = {}

#     def addSentence(self, sentence):
#         for word in sentence.split(' '):
#             self.addWord(word)

    def addWord(self, word):
        if word not in self.word2index:
            self.word2index[word] = self.n_words
            self.index2word[self.n_words] = word
            self.word2count[word] = 1
            self.n_words += 1
        else:
            self.word2count[word] += 1
            
        
```

Create a bag of words representation

```python
word2index = {}
tokenize = Lang()
for _ in chats:
    words_messages = [message.split(" ") for message in _]
    
    word_list = []
    for words in words_messages:
        word_list.extend(words)
    words_messages = word_list
    for char in ",.!?": # remove punctuation marks (count punctuation as same as words)
        words_messages = [message.split(char) for message in words_messages]
        words_messages = [item for sublist in words_messages for item in sublist]

    for char in "<-'":
        words_messages = [message.replace(char, "") for message in words_messages]
            
   
    words_messages = [message.lower() for message in words_messages]
    print (words_messages)
    for word in words_messages:
        tokenize.addWord(word)
    
```

```python
X = []
for _ in chats:
    words_messages = [message.split(" ") for message in _]
    
    word_list = []
    for words in words_messages:
        word_list.extend(words)
    words_messages = word_list
    for char in ",.!?":
        words_messages = [message.split(char) for message in words_messages]
        words_messages = [item for sublist in words_messages for item in sublist]

    for char in "<-'":
        words_messages = [message.replace(char, "") for message in words_messages]
            
   
    words_messages = [message.lower() for message in words_messages]

    bago_words = np.zeros(tokenize.n_words,)
    for word in words_messages:
        bago_words[tokenize.word2index[word]] += 1
    X.append(bago_words)
    
```

```python
X = np.array(X)
Y = np.array(fractures)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print (X_train.shape)
```

```python
from sklearn.model_selection import KFold, cross_val_score

for _ in range(1, 50, 10):
    k_fold = KFold(n_splits=5, shuffle=True)
    knn = KNeighborsClassifier(n_neighbors=_, weights='distance')
    print(_, cross_val_score(knn, X_train, y_train, cv=k_fold, n_jobs=-1))
```

```python
from sklearn.linear_model import LogisticRegression

k_fold = KFold(n_splits=5, shuffle=True)
lr = LogisticRegression()
print(_, cross_val_score(lr, X_train, y_train, cv=k_fold, n_jobs=-1))
```

```python
from sklearn import svm
for _ in np.linspace(0.1, 10, 5):
    sv = svm.SVC(kernel='rbf', C = _)
    scores = cross_val_score(sv, X_train, y_train, cv=k_fold, n_jobs=-1)
    print(_, scores, np.mean(scores))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000)
# rf.fit(X_train, y_train)
scores = cross_val_score(sv, X_train, y_train, cv=k_fold, n_jobs=-1)
# print(accuracy_score(rf.predict(X_test), y_test))
print (scores, np.mean(scores))
```
