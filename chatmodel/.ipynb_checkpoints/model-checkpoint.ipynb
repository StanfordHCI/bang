```python
import json
import ast 
import pandas as pd
import numpy as np
from IPython import display
data = []
with open('bang/chats', "r") as f:
    for line in f:
        data.append(ast.literal_eval(line))
with open('mark/chats', "r") as f:
    for line in f:
        data.append(ast.literal_eval(line))


```

```python
batches = (set([chat["batch"] for chat in data if "batch" in chat.keys()]))
```

```python
print (batches)
```

```python
for _ in batches:
    batch = [(chat["userID"], chat["message"], chat["room"], chat["time"]) 
                          for chat in data if ("batch" in chat.keys()) and (chat["batch"] == _)]
    batch = pd.DataFrame(batch)
    batch.columns = ["userID", "message", "team", "time"]
    batch = batch.sort_values(["team", "time"])
    batch.index = np.arange(1, batch.shape[0] + 1)
    display.display(batch)
    
```
